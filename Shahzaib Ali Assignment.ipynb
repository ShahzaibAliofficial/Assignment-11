```python
def sql_query(q):
    return pd.read_sql_query(q,conn)
```

```python
sql_query('select * from sqlite_master')
```

```python
q = 'select * from salaries'
df = sql_query(q)
df
```

### ***1. Show All Employess With A Total Pay Between 50,000 and 75,000***

```python
q = '''select EmployeeName,TotalPay from salaries
    where TotalPay between 50000 and 75000
'''
sql_query(q)
```

### ***2. Show All Employees With A BasePay less than 50,000 Or A TotalPay Greater Than 100,000***

```python
q = '''select EmployeeName,BasePay,TotalPay from salaries
    where BasePay < 50000
    and TotalPay > 100000
'''
sql_query(q)
```

### ***3. Show All Employees With A TotalPay Benefits Value Between 125,000 and 150,000 And A Job Title Containing Word 'Director'***

```python
q = '''select EmployeeName,TotalPayBenefits,JobTitle from salaries
     where TotalPayBenefits between 125000 and 150000
     and JobTitle like '%DIRECTOR%'
'''
sql_query(q)
```

### ***4. Show All Employees Ordered By Their TotalPay Benefits In Descending Order***

```python
q = '''select EmployeeName,TotalPayBenefits from salaries
    where TotalPayBenefits order by TotalPayBenefits desc
'''
sql_query(q)
```

### ***5. How many employees in the salaries table have OvertimePay equal to 0?***

```python
q = '''select count(*) as Total_Employee_with_0_OvertimePay from salaries
    where OvertimePay == 0
'''
sql_query(q)
```
